## Rule based cleaning

### Fix time- combine Year month and date into a datetime format

```python
flight_df = flight_df.with_columns(
    pl.concat_str(
        [pl.col("YEAR"), pl.lit("-"), pl.col("MONTH").cast(pl.Utf8), pl.lit("-"), pl.col("DAY").cast(pl.Utf8)]
    ).alias("DATE").str.strptime(pl.Date)
)

flight_df = flight_df.drop(['YEAR', 'MONTH', 'DAY'])
```

```python
flight_df.columns
```

```python
flight_df.head()
```

### change day of the week to weekdays

```python
day_dict = {1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday', 5: 'Friday', 6: 'Saturday', 7: 'Sunday'}
flight_df = flight_df.with_columns(pl.col("DAY_OF_WEEK").replace(day_dict))
```

```python
flight_df.head()
```

### Tail Number

```python
flight_df = flight_df.group_by('AIRLINE').map_groups(
    lambda df: df.with_columns(
        df['TAIL_NUMBER'].fill_null(strategy="forward"))
    )
```

```python
flight_df.null_count()
```

```python
cancellation_df = flight_df.drop
```

```python

def format_time(val):
    if val is None:
        return None
    elif val == 2400:
        val = 0
    val_str = f"{val:04d}"
    return time(int(val_str[0:2]), int(val_str[2:4]))

flight_df = flight_df.with_columns([
    pl.col('DATE'),
    pl.col('SCHEDULED_DEPARTURE').map_elements(format_time).alias('SCHEDULED_DEPARTURE_DATETIME'),
    pl.col('DEPARTURE_TIME').map_elements(format_time).alias('DEPARTURE_TIME_DATETIME'),
    pl.col('SCHEDULED_ARRIVAL').map_elements(format_time).alias('SCHEDULED_ARRIVAL_DATETIME'),
    pl.col('ARRIVAL_TIME').map_elements(format_time).alias('ARRIVAL_TIME_DATETIME')
])
flight_df.head(10)
```

# Feature Engineering- creating new features

DEPARTURE INDEX

```python
flight_df = flight_df.with_columns((pl.col('DEPARTURE_TIME') / pl.col('SCHEDULED_DEPARTURE')).alias('DEPARTURE_INDEX'))
flight_df.head()
```

ARRIVAL INDEX

```python
flight_df = flight_df.with_columns((pl.col('ARRIVAL_TIME') / pl.col('SCHEDULED_ARRIVAL')).alias('ARRIVAL_INDEX'))
```

Total Delay

```python
flight_df_new = flight_df.with_columns(pl.col("DEPARTURE_DELAY") + pl.col("ARRIVAL_DELAY").alias("TOTAL_DELAY"))

flight_df_new.head(10)
```

ROUTE

```python
flight_df = flight_df.with_columns((pl.col("ORIGIN_AIRPORT") + "-" + pl.col("DESTINATION_AIRPORT")).alias("ROUTE"))

flight_df.head()
```

TIME OF DAY_FOR_DEPARTURE AND ARRIVAL

```python
flight_df.columns
```

```python
def get_time_of_day(t):
    if t >= time(6, 0) and t <= time(11, 59):
        return "Morning"
    elif t >= time(12, 0) and t <= time(17, 59):
        return "Afternoon"
    elif t >= time(18, 0) and t <= time(23, 59):
        return "Evening"
    else:
        return "Night"

flight_df = flight_df.with_columns(
   pl.col('DEPARTURE_TIME_DATETIME').map_elements(get_time_of_day).alias('TIME_OF_DAY_FOR_DEPARTURE')
)
print(flight_df)
```

```python
flight_df = flight_df.with_columns(
   pl.col('ARRIVAL_TIME_DATETIME').map_elements(get_time_of_day).alias('TIME_OF_DAY_FOR_ARRIVAL')
)
print(flight_df)
```

SPEED IN AIRPLANE

```python
flight_df = flight_df.with_columns((pl.col('DISTANCE') / (pl.col('ELAPSED_TIME')/60)).alias('SPEED_OF_AIRPLANE'))
```

SEASONAL INDICATOR

```python
def get_season(date):
    month = date.month
    if 3 <= month <= 5:
        return "Spring"
    elif 6 <= month <= 8:
        return "Summer"
    elif 9 <= month <= 11:
        return "Autumn"
    else:
        return "Winter"

flight_df = flight_df.with_columns(
   pl.col('DATE').map_elements(get_season).alias('SEASON'),
)
print(flight_df)
```

AVERAGE TAXI DURATION BY AIRLINE

```python

flight_df = flight_df.with_columns((pl.col('TAXI_OUT') + pl.col('TAXI_IN')).alias('TOTAL_TAXI_DURATION'))

# Calculate the average taxi duration by airline
average_taxi_by_airline = flight_df.group_by('AIRLINE').agg([
    pl.col('TOTAL_TAXI_DURATION').mean().alias('AVERAGE_TAXI_DURATION')
])

average_taxi_by_airline.head(15)
```

PERCENTAGE OF FLGIHT DELAY FOR EACH AIRLINE

```python
df_delay = flight_df.with_columns(
    (pl.col('DEPARTURE_DELAY') > 0).cast(pl.UInt8).alias('IS_DELAYED')
)

percentage_delayed_by_airline = df_delay.group_by('AIRLINE').agg([
    (pl.col('IS_DELAYED').mean() * 100).alias('PERCENTAGE_DELAYED')
])

percentage_delayed_by_airline.head(15)
```

### Time Series Analysis

```python

monthly_flights = flight_df.group_by([pl.col('DATE').dt.year().alias('year'),pl.col('DATE').dt.month().alias('month')]).agg([pl.count().alias('num_flights'), pl.col('TOTAL_DELAY').sum().alias('total_delay_sum')
]).sort(['year', 'month'])

monthly_flights_pd = monthly_flights.to_pandas()
monthly_flights_pd['month_year'] = pd.to_datetime(monthly_flights_pd.assign(day=1)[['year', 'month', 'day']])

plt.figure(figsize=(12, 6))
plt.plot(monthly_flights_pd['month_year'], monthly_flights_pd['total_delay_sum'], marker='o')
plt.title('Total Delay Per Month Over Time')
plt.xlabel('Month of Order Date')
plt.ylabel('Total Delay (Sum)')
plt.grid(False)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

monthly_flights = flight_df.group_by(pl.col('DATE').dt.month().alias('month'), pl.col('DATE').dt.year().alias('year')).agg(pl.count().alias('num_flights')).sort(['year', 'month'])
monthly_flights_pd = monthly_flights.to_pandas()


monthly_flights_pd['month_year'] = pd.to_datetime(monthly_flights_pd.assign(day=1)[['year', 'month', 'day']])

plt.figure(figsize=(12, 6))
plt.plot(monthly_flights_pd['month_year'], monthly_flights_pd['num_flights'], marker='o')
plt.title('Number of Flights Per Month Over Time')
plt.xlabel('Month of Order Date')
plt.ylabel('Number of Flights')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
